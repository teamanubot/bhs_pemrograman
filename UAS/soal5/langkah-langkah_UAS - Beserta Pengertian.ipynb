## Pengertian Command

### 1. `cd UAS`

- **cd** : Perintah "change directory" digunakan untuk mengubah direktori kerja saat ini di command line.
- **UAS** : Nama direktori yang ingin Anda tuju.

**Fungsi** : Perintah ini mengubah direktori kerja saat ini ke direktori bernama `UAS`. Jika direktori `UAS` ada di dalam direktori saat ini, Anda akan berpindah ke dalamnya. Jika direktori `UAS` tidak ada, Anda akan mendapatkan pesan kesalahan.


### 2. `code .`

- **code** : Perintah ini digunakan untuk membuka Visual Studio Code dari command line. Agar perintah ini berfungsi, Anda harus memiliki Visual Studio Code terinstal dan `code` ditambahkan ke PATH sistem Anda.
- **.** : Tanda titik (.) menunjukkan direktori saat ini.

**Fungsi** : Perintah ini membuka Visual Studio Code dengan direktori kerja saat ini sebagai workspace. Artinya, semua file dan folder di dalam direktori saat ini akan ditampilkan di sidebar Visual Studio Code.

### 3. `docker-compose up -d --build`

- **docker-compose** : Perintah ini digunakan untuk menjalankan aplikasi multi-container menggunakan file `docker-compose.yml`.
- **up** : Perintah ini akan menjalankan kontainer yang didefinisikan di dalam file `docker-compose.yml`.
- **-d** : Opsi ini menjalankan kontainer di background (detached mode).
- **--build** : Opsi ini membangun ulang (rebuild) gambar Docker sebelum menjalankan kontainer, memastikan semua perubahan terbaru diterapkan.

**Fungsi** : Perintah ini akan membangun ulang gambar Docker dan menjalankan semua layanan yang ditentukan di dalam file `docker-compose.yml` di background.

### 4. `docker exec -it uas_pemrog bash`

- **docker exec** : Perintah ini digunakan untuk menjalankan perintah di dalam kontainer yang sedang berjalan.
- **-it** : Kombinasi opsi ini membuat terminal interaktif (`-i` untuk input interaktif, `-t` untuk mengalokasikan pseudo-TTY).
- **uas_pemrog** : Nama atau ID dari kontainer yang ingin Anda akses.
- **bash** : Perintah yang dijalankan di dalam kontainer; dalam hal ini, memulai shell `bash`.

**Fungsi** : Perintah ini membuka shell interaktif `bash` di dalam kontainer bernama `uas_pemrog`, memungkinkan Anda untuk menjalankan perintah secara langsung di dalam kontainer tersebut.

### 5. `composer create-project laravel/lumen .`

- **composer** : Manajer paket untuk PHP.
- **create-project** : Perintah ini digunakan untuk membuat proyek baru berdasarkan paket yang ditentukan.
- **laravel/lumen** : Nama paket yang akan diinstal, dalam hal ini framework Lumen.
- **.** : Direktori tempat proyek akan diinstal; dalam hal ini, direktori saat ini.

**Fungsi** : Perintah ini membuat proyek baru menggunakan framework Lumen di dalam direktori saat ini.

### 6. `composer require flipbox/lumen-generator`

- **composer** : Manajer paket untuk PHP.
- **require** : Perintah ini menambahkan dependensi baru ke proyek dan menginstalnya.
- **flipbox/lumen-generator** : Nama paket yang akan diinstal.

**Fungsi** : Perintah ini menambahkan dan menginstal paket `flipbox/lumen-generator` ke dalam proyek Lumen Anda, yang menyediakan generator artisan tambahan untuk Lumen.

### 7. `rm -rf ./composer.lock`

- **rm** : Perintah untuk menghapus file atau direktori.
- **-rf** : Opsi ini menghapus file dan direktori secara rekursif (`-r`) dan memaksa penghapusan (`-f`), tanpa meminta konfirmasi.
- **./composer.lock** : File yang akan dihapus; dalam hal ini, file `composer.lock` di direktori saat ini.

**Fungsi** : Perintah ini menghapus file `composer.lock`, yang berisi versi pasti dari semua paket yang diinstal. Menghapus file ini biasanya dilakukan untuk mengatasi masalah dependensi.

### 8. `mv .env.example .env`

- **mv** : Perintah untuk memindahkan atau mengganti nama file atau direktori.
- **.env.example** : Nama file sumber yang akan dipindahkan atau diganti namanya.
- **.env** : Nama file tujuan.

**Fungsi** : Perintah ini mengganti nama file `.env.example` menjadi `.env`. File `.env` digunakan untuk menyimpan konfigurasi lingkungan aplikasi.

### 9. `nano .env`

- **nano** : Editor teks berbasis terminal yang sederhana dan mudah digunakan.
- **.env** : Nama file yang akan diedit.

**Fungsi** : Perintah ini membuka file `.env` di editor teks `nano`, memungkinkan Anda untuk mengedit konfigurasi lingkungan aplikasi.

### 10. `php artisan key:generate`

- **php** : Interpreter untuk menjalankan skrip PHP.
- **artisan** : Tool command-line yang disediakan oleh Laravel untuk berbagai tugas terkait aplikasi.
- **key:generate** : Perintah artisan untuk menghasilkan kunci aplikasi yang baru.

**Fungsi** : Perintah ini menghasilkan kunci aplikasi baru dan memperbarui nilai `APP_KEY` di file `.env`. Kunci ini digunakan oleh Laravel untuk enkripsi, termasuk hashing password dan data sensitif lainnya.

### 11. `php artisan migrate`

- **php** : Interpreter untuk menjalankan skrip PHP.
- **artisan** : Tool command-line yang disediakan oleh Laravel untuk berbagai tugas terkait aplikasi.
- **migrate** : Perintah artisan untuk menjalankan migrasi database.

**Fungsi** : Perintah ini menjalankan semua migrasi yang belum dijalankan di database Anda. Migrasi digunakan untuk membuat dan memodifikasi tabel dan skema database.

### 12. `chmod 777 -R storage/*`

- **chmod** : Perintah untuk mengubah izin file atau direktori.
- **777** : Izin penuh untuk pemilik, grup, dan lainnya (baca, tulis, dan eksekusi).
- **-R** : Opsi ini menerapkan perubahan secara rekursif ke semua file dan subdirektori di dalam direktori yang ditentukan.
- **storage/** : Nama direktori yang izinnya akan diubah.

**Fungsi** : Perintah ini memberikan izin penuh (baca, tulis, eksekusi) untuk semua file dan subdirektori di dalam direktori `storage/`. Izin ini sering diperlukan agar Laravel dapat menulis file log dan cache.

### 13. `rm -rf app/Models/User.php`

- **rm** : Perintah untuk menghapus file atau direktori.
- **-rf** : Opsi ini menghapus file dan direktori secara rekursif (-r) dan memaksa penghapusan (-f), tanpa meminta konfirmasi.
- **app/Models/User.php** : Path ke file yang akan dihapus.

**Fungsi** : Perintah ini menghapus file `User.php` di dalam direktori `app/Models/`.

### 14. `php artisan make:Model User -mcs --resource`

- **php** : Interpreter untuk menjalankan skrip PHP.
- **artisan** : Tool command-line yang disediakan oleh Laravel untuk berbagai tugas terkait aplikasi.
- **make** : Perintah artisan untuk membuat model baru.
- **User** : Nama model yang akan dibuat.
- **-m** : Opsi untuk membuat file migrasi untuk model.
- **-c** : Opsi untuk membuat controller untuk model.
- **-s** : Opsi untuk membuat seeder untuk model.
- **--resource** : Opsi untuk membuat controller resource (dengan metode CRUD standar).

**Fungsi** : Perintah ini membuat model `User` lengkap dengan file migrasi, controller, dan seeder. Controller akan menggunakan metode resource standar.

### `15. php artisan migrate:fresh --seed`
- **php** : Interpreter untuk menjalankan skrip PHP.
- **artisan** : Tool command-line yang disediakan oleh Laravel untuk berbagai tugas terkait aplikasi.
- **migrate** : Perintah artisan untuk menghapus semua tabel dan menjalankan semua migrasi dari awal.
- **--seed** : Opsi untuk menjalankan seeder setelah migrasi selesai.

**Fungsi** : Perintah ini menghapus semua tabel dari database, menjalankan semua migrasi dari awal, dan kemudian menjalankan seeder untuk mengisi database dengan data awal.

## Laravel Lumen : Migrations, Controllers, dan Seeders

### **Migrations**
**Pengertian** <br>
Migrations adalah cara untuk mengelola skema basis data secara versioning dan terkontrol. Migrations memungkinkan pengembang untuk membuat, memodifikasi, dan menghapus tabel dan kolom dalam basis data menggunakan kode PHP, sehingga perubahan skema basis data bisa dilacak dan dikelola dengan baik. <br><br>

**Tujuan dan Manfaat**
- **Version Control** : Setiap perubahan pada skema basis data dicatat, memungkinkan pengembang untuk melacak perubahan dan kembali ke versi sebelumnya jika diperlukan.
- **Kolaborasi** : Memudahkan tim pengembang untuk bekerja bersama tanpa konflik pada perubahan basis data.
- **Automatisasi** : Migrations memungkinkan otomatisasi proses pengaturan skema basis data ketika meng-deploy aplikasi ke lingkungan yang berbeda.

### **Controllers**
**Pengertian** <br>
Controllers adalah bagian dari MVC (Model-View-Controller) yang bertanggung jawab untuk menangani logika permintaan HTTP, mengolah data melalui model, dan mengembalikan respon yang sesuai ke klien. Controllers dalam Lumen biasanya digunakan untuk mengorganisir logika aplikasi dalam satu tempat yang rapi. <br><br>

**Tujuan dan Manfaat**
- **Organisasi Kode** : Memisahkan logika bisnis dari routing, membuat kode lebih bersih dan mudah dipelihara.
- **Reuse** : Memungkinkan penggunaan kembali logika yang sama di berbagai bagian aplikasi.
- **Testability** : Mempermudah pengujian unit karena logika bisnis terpisah dari detail HTTP.

### **Seeders**
**Pengertian** <br>
Seeders adalah skrip yang digunakan untuk mengisi basis data dengan data awal atau data contoh. Mereka membantu pengembang untuk menginisialisasi basis data dengan data yang diperlukan untuk pengembangan dan pengujian. <br><br>

**Tujuan dan Manfaat**
- **Data Awal** : Mengisi tabel dengan data default yang dibutuhkan aplikasi.
- **Pengujian** : Menyediakan data contoh untuk pengujian fitur aplikasi.
- **Pengembangan** : Membuat lingkungan pengembangan yang siap pakai dengan data yang sesuai.

## **Kesimpulan**
- **Migrations** : Mengelola skema basis data secara terkontrol dan versioning.
- **Controllers** : Menangani logika permintaan HTTP dan mengembalikan respon yang sesuai.
- **Seeders** : Mengisi basis data dengan data awal atau contoh untuk pengembangan dan pengujian.
